The aim of this section is to group the data into analysable sections. 
Specifically the aim is to split the data into cases of the charge on both electrons being negative (ElectronX2) both electrons being positve (PositronX2) or a mix of one positive and one negative.

A bit of notes on my coding process.
I first segment my data based on the result and get mixed and mixed rows. mixed is purely the verification that the result has done what I wanted and Mixed is what will be used in further code.

I came into an issue with this as the orignal data frame was a numpy array, appending multiple rows into a numpy array results in a numpy array filled with several numpy arrays, one for each result. to bypass this I converted the rows into a list and then returned a numpy array.

I then repeat this for non mixed:pure.

I then further segmented pure into its two positve and two negative cases.

When Evaluating my code I occasionally printed a header beforehand which is just the variable names defined in the earlier section, before I printed the full row of data that was produced.

```python
#The Charges are neutralised if the charges on both electrons sum to zero:
print(""" Splitting Data frame into Either Neutralised 
(Net Charge of both Electrons is 0) or not Neutralised: """  )
#If the Sum of Charges is Zero then there is one positive and one negative or "mixed"
print('Charges Are Neutralised:')
#The Mixed variable is just to confirm that the the output only contains Q1,Q2 =(1,-1) or (-1,1)
#Sum zero:
def check_sum0():
    Mixed= []
    MixedRows=[]
    for i in range(len(df)):
        result = Q1[i]+Q2[i]
        row = df[i].tolist()
        strResult=str("[ ")+ str(Q1[i]) +str(' | ')+ str(Q2[i])+str(' ]')
        if result == 0:
            #here notice the i, causing an indexing element:
            Mixed.append((i,strResult))
            MixedRows.append(row)
    return (Mixed),np.array(MixedRows)

Mixed,MixedRows = check_sum0()
print(Mixed[:5])
print(sc)
print(Header)
print(MixedRows[:2])
print(sc)
#Good WORKS


#The Pure variable is just to confirm that the the output only contains (Q1,Q2)=(1,1) or (-1,-1)
print('Charge Not Neutralised:') 
def check_sumNot0():
    Pure= []
    PureRows=[]
    for i in range(len(df)):
        result = Q1[i]+Q2[i]
        row = df[i].tolist()
        strResult=str("[ ")+ str(Q1[i]) +str(' | ')+ str(Q2[i])+str(' ]')
        if result != 0:
            #here notice the i, causing an indexing element:
            Pure.append((i,strResult))
            PureRows.append(row)
    return Pure, np.array(PureRows)
Pure, PureRows = check_sumNot0()
print(Pure[:5])
print(sc)
print(Header)
print(PureRows[:3])
#Good WORKS


print(sc)

#There are two possible ways two charged particles can be not net 0. either both positive or both negative.
print('Splitting "Not Neutralised Charge" into the Positive and Negative Cases ')
def ElectronOPositron():
    ElectronX2= []
    PositronX2=[]
    for i in range(len(df)):
        row = df[i].tolist()  # Convert the row to a Python list
        if Q1[i]==1.0 and Q2[i]==1.0:
            PositronX2.append(row)
        elif Q1[i]==-1.0 and Q2[i]==-1.0:
            ElectronX2.append(row)
        else:
            pass
    return np.array(PositronX2), np.array(ElectronX2)
PositronX2, ElectronX2 = ElectronOPositron()


print('Two Positive Charges:')
print(PositronX2[:3])
print(sc)
print('Two Negative Charges:')
# for item in ElectronX2:
#     print(item)
print(ElectronX2[:3])
#YesGood
```

Now that the data has been grouped into analysable sections, the data may be analysed and the main <b> aim </b> of this project carried out. The main aim of this analysis is to build statistical evidence that supports or rejects what is already known about the Z boson.

A terminology Note:
Electrons are referred to as having either a positive or negative charge, if the charge is negative it is still called an electron but if it is positive it is called a positron. 

This is Summarised below:

A general Electron (charge - or +): reffered to as Electron I will rarely be talking about a single electron so this should never come up.

A specific electron (Charge - Only): Electron

A general pair of electrons (charge ++ or -- or -+ or +-): Electron pair.

A pair of Positrons (charge ++ only): Positron pair.

A pair of electrons(charge -- only): Electron pair

A pair of Electrons of opposite charges (charge (-+ or +-): Mixed pair

The Z boson is a mediatior of the weak force in that it provides the force given between subatiomic particles. The Z boson  exists for an extremely short time (10^-25s) hence the only way to observe it is to make it in the lab. After this time, the Z boson decays into a electronn electron positron pair. 
As a Z boson can only produce mixed pairs (Either +- or -+), there will be no evidence of a Z boson in the pure cases (++, --). 
The Z boson has a known mass of 91.3MeV. If the Z boson decays into a mixed pair, the invariant mass (mass of both electrons at rest) should be equal to the mass of the Z boson. 
Hence statistical evidence of the z boson having decayed could be contructed by a spike in the occurence of an electron positron pair for the invariant mass of the electron pair being the mass of the Z boson. The spike would be in referance to the pure cases of either electron pair (ElectronX2) or positron pair (PositronX2) in which the Z boson cant decay into.

```python
from matplotlib.colors import LogNorm
#The mathematical formula for the transverse energy is as produced below:
def TransverseEnergy1 (i):
    return np.tanh(eta1[i])*E1[i]
def TransverseEnergy2(i):
    return np.tanh(eta2[i])*E2[i]
#Putting all the data back into a numpy array:    
tEnergy1 = np.array([TransverseEnergy1(i) for i in range(len(eta1))])
tEnergy2 = np.array([TransverseEnergy2(i) for i in range(len(eta2))])

#we want it's absolute value of transverse energy and eta:
tEnergy1=np.abs(tEnergy1)
eta1=np.abs(eta1)
tEnergy2=np.abs(tEnergy2)
eta2=np.abs(eta2)

#Creation of subplots:
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
#Using LogNorm to be create more detail between bin edges.

hist1=ax1.hist2d(tEnergy1,eta1,norm=LogNorm() )
ax1.set_xlabel('Transverse Energy 1')
ax1.set_ylabel('Eta 1')

hist2=ax2.hist2d(tEnergy2,eta2,norm=LogNorm() )
ax1.set_xlabel('Transverse Energy 2')
ax1.set_ylabel("Eta 2")

fig.suptitle('2D Histograms of Transverse Energy and Eta')
```

Firstly lets understand the variables,
eta: called the pseudorapicity is a positional resprentation of where the event has taken place with respect to the position of the electron beam. 
Transverse energy on the otherhand is a variable repreesnting the how many energy of the electron is perpendicular to the beam axis. The transverse plane represents this perpendicular plane.

The graph correlation heat map demonstrates that the lowest transverse energy occurs with a low eta value. This means that increasing the eta value causes a larger component of the electrons momentum to lie in the transverse plane.

When eta is near 0, the electrons are almost paralled to the beam axis then the energy is almost paralled to the beam axis and hence there is no energy perpendicular to the beam axis.
Contrarily, if eta is increased then then a larger component of the electrons energy is in the transverse plane.

Hence the experimental results produced in the correlation heat map support this phenomenon.
